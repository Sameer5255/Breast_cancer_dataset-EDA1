## Problem Statement

The goal of this Exploratory Data Analysis (EDA) project is to explore and understand the Breast Cancer Wisconsin (Diagnostic) dataset. This dataset contains features computed from digitized images of fine needle aspirate (FNA) of breast masses, and each record is labeled as either benign (B) or malignant (M). The primary objective is to perform a detailed statistical and visual analysis to:

Understand the distribution of classes (benign vs malignant).

Identify which features contribute most to the diagnosis.

Detect patterns, correlations, and potential outliers within the data.

## Count plot to check which diagnosis is highest malignant or benign

```python

sns.countplot(data=data, x='diagnosis', palette='Set2')
plt.title('Diagnosis Distribution')
plt.xlabel('Diagnosis (B = Benign, M = Malignant)')
plt.ylabel('Count')
plt.show()
```

## Benign type of diagnisis is highest

##diagnosis respected to radius mean, texture mean, perimeter mean

```python
features = ['radius_mean', 'texture_mean', 'perimeter_mean']
for feature in features:
    plt.figure(figsize=(6,4))
    sns.boxplot(data=data, x='diagnosis', y=feature)
    plt.title(f'{feature} by Diagnosis')
    plt.show()
```

## in all three we observed mmalignant type of diagnosis is more common and  highest outliers present in begign type of texture_mean.

##check correlation in between all the column

```python
df_corr = data.drop(columns=['id', 'diagnosis', 'Unnamed: 32'], errors='ignore')
corr = df_corr.corr()

plt.figure(figsize=(14,12))
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt='.2f', square=True, cbar=True)
plt.title('Feature Correlation Heatmap')
plt.show()
```

##Conclusion

The exploratory data analysis of the breast cancer dataset reveals that the Benign (B) diagnosis is the most common overall. However, when focusing on key features such as radius_mean, texture_mean, and perimeter_mean, the Malignant (M) diagnosis tends to dominate, indicating these features are important markers for malignancy.

Additionally, the feature texture_mean shows the highest number of outliers within the benign cases, suggesting greater variability in this measurement among non-cancerous tumors.

Finally, the correlation analysis indicates no strong linear relationships between the features, which highlights the complexity of the data and suggests that a combination of multiple features will likely be necessary for effective classification.
